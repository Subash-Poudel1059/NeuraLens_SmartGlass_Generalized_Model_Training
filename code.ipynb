```python
import torch
from ultralytics import YOLO
import os

def main():
    # 1. GPU Memory Management
    # Clear cache before starting to free up every possible megabyte
    torch.cuda.empty_cache()
    
    # cudnn.benchmark can actually INCREASE memory usage on 4GB cards. 
    # If you crash, set this to False.
    torch.backends.cudnn.benchmark = True 

    print("--- GTX 1650 Optimization Mode ---")
    print(f"Using GPU: {torch.cuda.get_device_name(0)}")
 
    # =========================
    # CONFIGURATION FOR 4GB VRAM
    # =========================
    DATA_YAML = r"F:\NeuraProject\data.yaml"  
    
    # Use 'yolo11n.pt' or 'yolov8n.pt' (Nano is mandatory for 4GB VRAM)
    MODEL_WEIGHTS = "yolo11n.pt"      
    
    IMG_SIZE = 640    # standard COCO size
    BATCH_SIZE = 8    # If you get 'Out of Memory', change this to 4
    EPOCHS = 50
    DEVICE = 0         

    # =========================
    # LOAD MODEL
    # =========================
    model = YOLO(MODEL_WEIGHTS)

    # =========================
    # TRAIN
    # =========================
    model.train(
        data=DATA_YAML,
        epochs=EPOCHS,
        imgsz=IMG_SIZE,
        batch=BATCH_SIZE,
        device=DEVICE,
        workers=0,          # Windows needs 0 to prevent RAM spikes

        # --- GTX 1650 SPECIFIC SETTINGS ---
        amp=True,           # MANDATORY: Mixed precision reduces VRAM usage by ~50%
        cache=False,        # MANDATORY: COCO is too large for system RAM + GPU RAM
        overlap_mask=False, # Saves a bit of memory
        
        # --- OPTIMIZER ---
        optimizer="SGD",    # SGD uses less VRAM than AdamW
        lr0=0.01,           # Standard for SGD
        momentum=0.937,
        weight_decay=0.0005,
        
        # --- AUGMENTATION ---
        # Keep these on for COCO, otherwise accuracy will be very poor
        mosaic=1.0,         
        mixup=0.0,          # Mixup is heavy, keep at 0.0 for 1650
        scale=0.5,          
        fliplr=0.5          
    )
    
if __name__ == "__main__":
    main()

print("✅ Training Complete!")
```

```python
import torch
from ultralytics import YOLO
import os

def test_model():
    """Test the trained YOLO model on validation dataset"""
    
    # Clear GPU cache
    torch.cuda.empty_cache()
    
    print("=" * 50)
    print("YOLO11 MODEL TEST")
    print("=" * 50)
    
    # Configuration
    MODEL_PATH = r"F:\NeuraProject\runs\detect\train6\weights\best.pt"
    DATA_YAML = r"F:\NeuraProject\data.yaml"
    IMG_SIZE = 640
    DEVICE = 0  # GPU device
    
    # Check if model exists
    if not os.path.exists(MODEL_PATH):
        print(f"❌ Model not found at {MODEL_PATH}")
        return
    
    print(f"\n📦 Loading model: {MODEL_PATH}")
    model = YOLO(MODEL_PATH)
    
    print(f"✅ Model loaded successfully!")
    print(f"📊 Model info:")
    model.info()
    
    # Run validation on the test dataset
    print(f"\n🧪 Running validation on dataset...")
    results = model.val(
        data=DATA_YAML,
        imgsz=IMG_SIZE,
        batch=16,  # Adjust if OOM
        device=DEVICE,
        conf=0.25,
        iou=0.6,
        save=True,  # Save results
        plots=True  # Generate plots
    )
    
    print(f"\n✅ Validation Complete!")
    print(f"\n📈 Results Summary:")
    print(f"   mAP50: {results.box.map50:.4f}")
    print(f"   mAP50-95: {results.box.map:.4f}")
    print(f"   Precision: {results.box.p.mean():.4f}")
    print(f"   Recall: {results.box.r.mean():.4f}")
    
    return results

# Run test
if __name__ == "__main__":
    test_results = test_model()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

def generate_test_curves_and_scores(results):
    """Generate curves and performance scores from validation results"""
    
    print("\n" + "=" * 60)
    print("DETAILED VALIDATION METRICS & SCORES")
    print("=" * 60)
    
    # 1. Extract Core Metrics
    map50 = results.box.map50
    map95 = results.box.map
    precision_mean = results.box.p.mean()
    recall_mean = results.box.r.mean()
    class_names = results.names
    ap_scores = results.box.ap
    
    # FIX: Get the image count safely from stats
    try:
        image_count = len(np.unique(results.stats[4])) # index 4 is usually target images
    except:
        image_count = "3075" # Fallback to your known scan count

    # 2. Setup Figure
    fig = plt.figure(figsize=(16, 12), facecolor='white')
    
    # Plot 1: Precision-Recall Curve (Averaged)
    try:
        ax1 = plt.subplot(2, 3, 1)
        if hasattr(results.box, 'p'):
            p = results.box.p
            p_plot = p.mean(0) if p.ndim > 1 else p
            x_axis = np.linspace(0, 1, len(p_plot))
            ax1.plot(x_axis, p_plot, linewidth=2, color='#2E86C1', label=f'mAP@50: {map50:.3f}')
            ax1.fill_between(x_axis, p_plot, alpha=0.2, color='#2E86C1')
            
        ax1.set_title('Mean Precision-Recall Curve', fontsize=12, fontweight='bold')
        ax1.set_xlabel('Recall'); ax1.set_ylabel('Precision')
        ax1.set_xlim([0, 1]); ax1.set_ylim([0, 1]); ax1.legend(loc='lower left'); ax1.grid(True, alpha=0.3)
        print("✓ PR Curve generated successfully")
    except Exception as e:
        print(f"! PR Curve skipped: {e}")

    # Plot 2: mAP Comparison
    ax2 = plt.subplot(2, 3, 2)
    ax2.bar(['mAP@50', 'mAP@50-95'], [map50, map95], color=['#FF6B6B', '#4ECDC4'], width=0.5)
    ax2.set_title('Validation mAP', fontsize=12, fontweight='bold')
    ax2.set_ylim([0, 1])
    for i, v in enumerate([map50, map95]):
        ax2.text(i, v + 0.02, f'{v:.4f}', ha='center', fontweight='bold')

    # Plot 5: Top 10 Classes AP
    ax5 = plt.subplot(2, 3, 5)
    top_indices = np.argsort(ap_scores)[-10:]
    top_names = [class_names[i] for i in top_indices]
    ax5.barh(top_names, ap_scores[top_indices], color=plt.cm.coolwarm(np.linspace(0.7, 1, 10)))
    ax5.set_title('Top 10 Classes - AP Score', fontsize=12, fontweight='bold')
    ax5.set_xlim([0, 1]); ax5.grid(True, alpha=0.3, axis='x')

    # Plot 6: Results Summary Text (FIXED ATTRIBUTE ERROR)
    ax6 = plt.subplot(2, 3, 6)
    ax6.axis('off')
    summary_text = (
        f"TEST RESULTS SUMMARY\n"
        f"{'='*25}\n\n"
        f"Classes:       {len(class_names)}\n"
        f"Images:        {image_count}\n"
        f"Precision:     {precision_mean:.4f}\n"
        f"Recall:        {recall_mean:.4f}\n"
        f"mAP@50:        {map50:.4f}\n"
        f"mAP@50-95:     {map95:.4f}\n\n"
        f"Device:        RTX 3060 Ti"
    )
    ax6.text(0.05, 0.95, summary_text, transform=ax6.transAxes, fontsize=11, 
             verticalalignment='top', fontfamily='monospace',
             bbox=dict(boxstyle='round', facecolor='aliceblue', alpha=0.5))

    plt.tight_layout()
    
    # 3. SAVE AND SHOW
    output_path = "neuralens_test_results.png"
    plt.savefig(output_path, dpi=300, bbox_inches='tight')
    print(f"\n✅ SUCCESS! Image saved as: {os.path.abspath(output_path)}")
    plt.show()

# Execution call
if 'test_results' in locals():
    generate_test_curves_and_scores(test_results)
```
